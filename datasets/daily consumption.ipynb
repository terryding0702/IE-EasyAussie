```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#clothing and footware
res = requests.get("https://liveinmelbourne.vic.gov.au/live/money-and-tax/cost-of-living-in-melbourne/melb-china-cost-of-clothing-and-footwear")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')
df_cloth = pd.read_html(str(table))
df_cloth[3]
```
